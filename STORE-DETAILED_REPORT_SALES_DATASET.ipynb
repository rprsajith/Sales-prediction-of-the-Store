```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_customer = data.groupby('CLIENT').sum().sort_values('AMOUNT_AFTER_TAX', ascending = False)[['AMOUNT_AFTER_TAX']][0:20]
sns.barplot(x = 'AMOUNT_AFTER_TAX',y = top_customer.index, data=top_customer)
plt.figure(figsize = (500,500))

#this code for which customer have buy more amount in restaurant with groupby
```

```python
top_tax = data.groupby('PRODUCT').sum().sort_values('TAXABLE', ascending = False)[['TAXABLE']][0:30]
sns.barplot(x = 'TAXABLE',y = top_tax.index, data = top_tax)

#this code for find which product highly taxed
```

```python
top_product = data.groupby('PRODUCT').sum().sort_values('QUANTITY', ascending = False)[['QUANTITY']][0:30]
sns.barplot(x = 'QUANTITY', y = top_product.index, data = top_product)

#this code for find which product is buy more
```

```python
from wordcloud import WordCloud
```

```python
text = ' '.join(data['PRODUCT'].dropna().astype(str).values)
```

```python
#create a wordcloud object

wordcloud = WordCloud(background_color="white", width = 800, height = 400).generate(text)
```

```python
#create a matplotlib figure and axis

fig,ax = plt.subplots(figsize=(10, 6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_title('word cloud of product')
ax.set_xticks([])
ax.set_yticks([])
plt.show

#display the word cloud on axis
```
